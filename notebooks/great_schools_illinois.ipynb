```python
import requests
import pandas as pd
import time
import os
os.chdir(os.path.dirname(os.getcwd()))
```

```python
STATE = 'IL'

# Get a list of all zip codes in the ilinois state
zip_codes = pd.read_csv('data/geo/zip_codes.csv')
zip_codes = zip_codes[zip_codes['state'] == STATE]
zip_codes_lst = zip_codes['zip'].tolist()
```

```python
API_KEY = 'kICbx7Ja0naaVKCG0y4dG6sT86dWH1vA5UdyG2YI'

def fetch_schools_by_zip(zip, level_codes='m'):
    """
    Fetch schools by zip code and level codes

    Input:
      zip: str, zip code
      level_codes: str, level codes, use '|' to separate multiple codes
    Output:
      json object containing information about schools and links
    """
    
    query_params = f'zip={zip}&level_codes={level_codes}&limit=50'
    query_base = "https://gs-api.greatschools.org/v2/schools?"

    query_url = query_base + query_params

    payload={}
    headers = {
      'Accept': 'application/json',
      'Content': 'application/json',
      'X-API-Key': API_KEY
    }

    response = requests.request("GET", query_url, headers=headers, data=payload)

    if response.json().get('schools') == []:
        return None

    return response.json()

```

```python
schools = pd.DataFrame()

for zip in zip_codes_lst:
    response = fetch_schools_by_zip(zip)
    time.sleep(0.1)
    if response is not None:
        zip_school = pd.DataFrame(response['schools'])
        zip_school['zip'] = zip
        schools = pd.concat([schools, zip_school], ignore_index=True)
        
```
